# Matrix Operations:

For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:

1. Prove A.A−1 = I.

```python
import numpy as np

# Define matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
A_inv = np.linalg.inv(A)

# Calculate the product of A and its inverse
identity_matrix = np.dot(A, A_inv)

# Print the result
print(identity_matrix)
```

2. Prove AB ̸= BA.

```python
import numpy as np

# Define matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
# Calculate AB
AB = np.dot(A, B)

# Calculate BA
BA = np.dot(B, A)

# Print the results
print("AB:")
print(AB)
print("\nBA:")
print(BA)

# Check if AB is not equal to BA
print("\nIs AB not equal to BA?", not np.array_equal(AB, BA))
```

3. Prove (AB)

T = BTAT

```python
import numpy as np

# Define matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
# Calculate (AB)T
AB_transpose = np.transpose(np.dot(A, B))

# Calculate BTAT
B_transpose_A_transpose = np.dot(np.transpose(B), np.transpose(A))

# Print the results
print("(AB)T:")
print(AB_transpose)
print("\nBTAT:")
print(B_transpose_A_transpose)

# Check if (AB)T is equal to BTAT
print("\nIs (AB)T equal to BTAT?", np.array_equal(AB_transpose, B_transpose_A_transpose))
```

# Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}\

```python
import numpy as np

# Coefficient matrix
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

# Constant terms
B = np.array([-1, -3, 9])

# Calculate the inverse of A
A_inv = np.linalg.inv(A)

# Solve for X by multiplying A_inv with B
X = np.dot(A_inv, B)

# Print the solution
print("Solution:")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])
```

# Experiment: How Fast is Numpy?

1. Element-wise Addition:
• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure
and Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for
this operation.

```python
import time
import numpy as np

# Create two lists of size 1,000,000
list1 = list(range(1000000))
list2 = list(range(1000000))

# Element-wise addition using Python lists
start_time = time.time()
result_list = [x + y for x, y in zip(list1, list2)]
end_time = time.time()
list_time = end_time - start_time
print(f"Time taken using Python lists: {list_time:.6f} seconds")

# Create two NumPy arrays of size 1,000,000
array1 = np.arange(1000000)
array2 = np.arange(1000000)

# Element-wise addition using NumPy arrays
start_time = time.time()
result_array = array1 + array2
end_time = time.time()
array_time = end_time - start_time
print(f"Time taken using NumPy arrays: {array_time:.6f} seconds")

# Compare the execution times
print(f"\nNumPy is {list_time / array_time:.2f} times faster than Python lists for this operation.")
```

2. Element-wise Multiplication
• Using Python Lists, perform element-wise multiplication of two lists of size 1, 000, 000.
Measure and Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for

```python
import time
import numpy as np

# Create two lists of size 1,000,000
list1 = list(range(1000000))
list2 = list(range(1000000))

# Element-wise multiplication using Python lists
start_time = time.time()
result_list = [x * y for x, y in zip(list1, list2)]
end_time = time.time()
list_time = end_time - start_time
print(f"Time taken using Python lists: {list_time:.6f} seconds")

# Create two NumPy arrays of size 1,000,000
array1 = np.arange(1000000)
array2 = np.arange(1000000)

# Element-wise multiplication using NumPy arrays
start_time = time.time()
result_array = array1 * array2
end_time = time.time()
array_time = end_time - start_time
print(f"Time taken using NumPy arrays: {array_time:.6f} seconds")

# Compare the execution times
print(f"\nNumPy is {list_time / array_time:.2f} times faster than Python lists for this operation.")
```

• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and
Print the time taken for this operation.

```python
import time
import numpy as np

# Create two lists of size 1,000,000
list1 = list(range(1000000))
list2 = list(range(1000000))

# Dot product using Python lists
start_time = time.time()
result_list = sum(x * y for x, y in zip(list1, list2))
end_time = time.time()
list_time = end_time - start_time
print(f"Time taken using Python lists: {list_time:.6f} seconds")

# Create two NumPy arrays of size 1,000,000
array1 = np.arange(1000000)
array2 = np.arange(1000000)

# Dot product using NumPy arrays
start_time = time.time()
result_array = np.dot(array1, array2)
end_time = time.time()
array_time = end_time - start_time
print(f"Time taken using NumPy arrays: {array_time:.6f} seconds")

# Compare the execution times
print(f"\nNumPy is {list_time / array_time:.2f} times faster than Python lists for this operation.")
```
